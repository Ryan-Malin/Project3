```python
def clean_labels(x):
    newcols = {}
    for thing in range(len(x.columns)):
        newthing = x.columns[thing].replace("_", " ").replace("num ", "number ").title()
        newcols[x.columns[thing]] = newthing
    return x.rename(columns = newcols)
```

```python
mhealth = [
    'average_number_of_mentally_unhealthy_days',
    'mental_health_provider_rate',
    'num_mental_health_providers',
    'percent_frequent_mental_distress'
]
```

```python
top_categories = [
    "percentile_rank_no_highschool_diploma",
    "percent_frequent_physical_distress",
    "percent_smokers",
    "percent_fair_or_poor_health",
    "percent_below_poverty",
    "percentile_rank_below_poverty",
    "percentile_rank_per_capita_income",
    "median_household_income",
    "percentile_rank_no_highschool_diploma",
    "percent_frequent_physical_distress",
    "average_number_of_physically_unhealthy_days",
    "percent_smokers",
    "percent_fair_or_poor_health",
    "percent_below_poverty",
    "percentile_rank_per_capita_income",
    "percentile_rank_social_vulnerability",
    "per_capita_income"
]
```

```python
geo = [
    'fips',
    'state',
    'county',
    'lat',
    'lon'
]
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../us_county_sociohealth_data.csv")
```

```python
# data for North Carolina counties
# dropped hiv columns, which had only two counties reporting
nc_df = df[df["state"] == "North Carolina"].drop(columns = ["hiv_prevalence_rate", "num_hiv_cases"])
```

```python
mhealth_df = nc_df[sorted(mhealth)]
geo_df = nc_df[geo]
top_categories_df = nc_df[sorted(set(top_categories))]
other_df = nc_df.drop(columns = mhealth).drop(columns = geo).drop(columns = top_categories)
other_df = other_df[sorted(list(other_df.columns))]
```

```python
clean_labels(nc_df).to_json("json/rawdata.json")
```
