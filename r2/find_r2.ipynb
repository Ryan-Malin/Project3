```python
import pandas as pd
import scipy.stats as st
```

```python
df = pd.read_csv("../us_county_sociohealth_data.csv")
```

```python
nc_df = df.loc[df["state"] == "North Carolina"]
nc_df.head()
```

```python
column_list = [
    'average_number_of_mentally_unhealthy_days',
    'num_mental_health_providers',
    'mental_health_provider_rate',
    'percent_frequent_mental_distress'
]
```

```python
# column range
my_range = [121, 160]

corr_df = pd.DataFrame(columns = ["x", "y", "r", "r2"])

# loop through mental health categories
for x in range(4):
    print("---------------------------")
    print(column_list[x])
# loop through categories in selected column range
    for y in range(my_range[0], my_range[1] + 1):
# compare different columns with number values
        if column_list[x] != nc_df.columns[y] and nc_df[nc_df.columns[y]].dtypes != "object":
# ignore row if either column is NaN
            new_df = nc_df[[column_list[x], nc_df.columns[y]]].dropna(axis=0,how="any")
# calculate r value
# x_values = mental health column
# y_values = column to compare
            x_values = new_df[new_df.columns[0]]
            y_values = new_df[new_df.columns[1]]
            r = st.pearsonr(x_values, y_values)[0]
            r2 = r ** 2
# save to dataframe
            corr_df = corr_df.append([{
                "x": new_df.columns[0],
                "y": new_df.columns[1],
                "r": r.round(4),
                "r2": r2.round(4)
            }], ignore_index = True)
```
