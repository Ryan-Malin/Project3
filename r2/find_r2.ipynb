```python
import pandas as pd
import scipy.stats as st
```

```python
df = pd.read_csv("us_county_sociohealth_data.csv")
```

```python
my_columns = list(df.columns[121:161])

column_list = [
    'average_number_of_mentally_unhealthy_days',
    'num_mental_health_providers',
    'mental_health_provider_rate',
    'percent_frequent_mental_distress'
]

column_list.extend(my_columns)
#column_list
```

```python
nc_df = df.loc[df["state"] == "North Carolina"][column_list]
nc_df.head()
```

```python
for x in range(4):
    print(nc_df.columns[x])
```

```python
corr_df = pd.DataFrame(columns = ["x", "y", "r", "r2"])

for x in range(4):
    print("---------------------------")
    print(nc_df.columns[x])
    for y in range(4, len(nc_df.columns)):
        new_df = nc_df[[nc_df.columns[x], nc_df.columns[y]]].dropna(axis=0,how="any")
        x_values = new_df[new_df.columns[0]]
        y_values = new_df[new_df.columns[1]]
        r = st.pearsonr(x_values, y_values)[0]
        r2 = r ** 2

        corr_df = corr_df.append([{
            "x": new_df.columns[0],
            "y": new_df.columns[1],
            "r": r.round(4),
            "r2": r2.round(4)
        }], ignore_index = True)
```
